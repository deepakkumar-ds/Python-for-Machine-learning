```python
##write a program to calculate simple interest using the formula (SI = P × R × T / 100).

P=float(input("Enter the principal amount:"))
R=float(input("Enter the Rate of interest:"))
T=float(input("Enter the Time:"))

SI = (P*R*T)/100
print("Simple Interest amount:",SI)
```

```python
##Write a program where the user enters Red, Yellow, or Green, and the program displays the corresponding action or "Invalid color."

color=input("light color:")
color = color.lower()

if (color=="red"):
    print("Stop")
elif(color=="yellow"):
    print("Ready to move")
elif(color=="green"):
    print("Go")
else:
    print("invalid Color")
```

```python
##Write a program that takes an integer and checks whether it is odd or even and whether it is divisible by 5.

num =int(input("Enter the number:"))

if(num%2==0):
    print(f"{num} is even")
else:
    print(f"{num} is odd")


if (num%5==0):
    print(f"{num} is divisble by 5")
else:
    print(f"{num} is not divisble by 5")
```

```python
##Write a program that checks password strength as Strong (≥8 characters), Medium (5–7 characters), or Weak (<5 characters).

password = input("Enter your password:")

length = len(password)

if (length>=8):
    print("Strong password")
elif (length>5 and length<7):
    print("Medium")
else:
    print("weak password")
```

```python
##Write a program that accepts electricity units consumed and calculates the total bill using the slab rates.
units=int(input("Enter units:"))

if(units<=100):
    amount=units*0.5

elif(units>100 and units<200):
    amount=(100*0.5)+(units-100)*0.75

else:
    amount=(100*0.5)+(100*0.75)+(units-200)*1

print("n/-----EB bill----")
print(f"units:{units}")
print(f"amount:{amount}")
```
